# Plotting demos

```python
from matplotlib import pyplot as plt
```

```python
xs = list(range(-10,11))
ys = [x*x for x in xs]
plt.plot(xs,ys,label="y = x^2")
plt.plot(xs,[x**3 for x in xs],label="y = x^3")
plt.grid()
plt.title("parabola y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
xs = list(range(-10,11))
ys = [x*x for x in xs]
plt.hist(ys,label="y = x^2")
plt.grid()
plt.title("parabola y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
